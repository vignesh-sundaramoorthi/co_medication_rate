```python
#Set up dependencies
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
#file to load
file_to_load = "Data/byetta_insulin_music_diagram_v5.csv"

df = pd.read_csv(file_to_load)
df.head()
```

```python
new_df = pd.melt(df, id_vars=['Patient ID'], var_name='date')
```

```python
mapping = {'NaN': 0, 'By': 2, 'In': 6}
new_df = new_df.replace({'value': mapping})
```

```python
new_df = new_df.replace(np.nan, 0)
```

```python
new_df.head()
```

```python
new_df['value_b'] = new_df['value']
new_df['value_in'] = new_df['value']
```

```python
del new_df['value']
```

```python
new_df['value_b'][new_df['value_b'] == 6]= 0
new_df['value_in'][new_df['value_in'] == 2]= 0
```

```python
new_df.head()
```

```python
df3 = new_df.set_index(['Patient ID', 'date']).sum(level=[0,1])
df3.reset_index()
```

```python
df3.to_csv('test', encoding='utf-8', index=True)
```
